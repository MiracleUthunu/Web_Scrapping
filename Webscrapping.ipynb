```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_manufacturing_companies_by_revenue'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')
```

```python
soup.find_all('table')[1]
```

```python
soup.find_all('table', class_ = "wikitable sortable plainrowheads")
```

```python
table = soup.find_all('table')[1]
print(table)
```

```python
 uk_titles = table.find_all('th')
```

```python
 uk_titles
```

```python
uk_titles_table = [title.text.strip() for title in uk_titles]
print(uk_titles_table)
```

```python
#put into pandas dataframe
df = pd.DataFrame(columns=uk_titles_table)
df
```

```python
column_data = table.find_all('tr')
column_data
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    length = len(df )
    df.loc[length] = individual_row_data
```

```python
df
```

```python
df.to_csv(r'scraped_data.csv', index=False)
```
